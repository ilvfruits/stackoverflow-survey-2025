```python
import os
import shutil

def copy_figure(SRC_DIR, DST_DIR):
    """
    This function does not return any value. It copies all figures from source dirctory to destine directory.
    
    Input param SRC_DIR: Dirtory containing the original figures.
    Input param DST_DIR: Dirtory to copy figures to.
    """
    os.makedirs(DST_DIR, exist_ok=True)

    copied = 0

    for file in os.listdir(SRC_DIR):
        if file.lower().endswith((".png", ".jpg", ".jpeg", ".svg")):
            src = os.path.join(SRC_DIR, file)
            dst = os.path.join(DST_DIR, file)
            shutil.copy2(src, dst)
            copied += 1
    
    print(f"Copied {copied} image(s) from {SRC_DIR} → {DST_DIR}")
```

```python
SRC_DIR = "../../reports/figures"
DST_DIR = "../../docs/assets/images"
copy_figure(SRC_DIR, DST_DIR)
```
